**●	Handle missing values appropriately (e.g., imputation, deletion).**

```python
FHR.isnull().sum()
```

deletion of null data might loose important patterns so i am using statistical methods like mean and median
based on the distribution null tuples are replaced with mean or median

```python
FHR_columns = ['LB',	'AC',	'DS',	'DP',	'MLTV',	'Width',	'Tendency',	'NSP']
for col in FHR_columns:
  median = print(f'median of {col}',FHR[col].median())
  mean = print(f'mean of {col}',FHR[col].mean())
  if(mean==median):
    FHR[col].fillna(FHR[col].mean(),inplace=True)
  else:
    FHR[col].fillna(FHR[col].median(),inplace=True)
```

```python
FHR.isnull().sum()
```

the missing values are been replaced

**●	Detect and treat outliers if necessary.**

```python
FHR_column = ['LB',	'AC','FM','UC','DL','DS',	'DP','ASTV','MSTV','ALTV','MLTV',	'Width',	'Tendency',	'NSP']
for quar in FHR_column:
  Q1 = FHR[quar].quantile(0.25)
  Q3 = FHR[quar].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = FHR[(FHR[quar] < lower_bound) | (FHR[quar] > upper_bound)]
  len_outliers = len(outliers)
  print(f"Number of outliers in {quar}: {len_outliers}")
```

we have found out liers and there are methods to treat out liers but in this case out liers are the actual reading of fetal hart rate so changing in outliers might cause lot of changes so i am leaving them as they are

# 2.	Statistical Summary:

** ●	Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median) and dispersion (standard deviation, interquartile range).**

```python
for stat in FHR_column:
  print(f'mean of {stat}',FHR[stat].mean())
  print(f'median of {stat}',FHR[stat].median())
  print(f'standard deviation of {stat}',FHR[stat].std())
  print(f'interquartile range of {stat}',FHR[stat].quantile(0.75)-FHR[stat].quantile(0.25))
  print('---------------------------------------')
```

**●	Highlight any interesting findings from this summary.**

 Mean vs. Median (Skewness)
If mean > median → data is right-skewed (long tail on the right).

If mean < median → data is left-skewed (long tail on the left).

If mean ≈ median → data is likely symmetrical.

What to say:

"The mean is noticeably higher than the median, suggesting the presence of a right-skewed distribution possibly caused by high outliers."

2. Standard Deviation (Spread Around the Mean)
A high std indicates high variability; data points are far from the mean.

A low std means the data is closely packed around the mean.

What to say:

"A relatively high standard deviation indicates significant variation in the data, which may impact the consistency of results."

3. Interquartile Range (IQR – Spread of Middle 50%)
Large IQR: Data is spread out even in the middle 50%.

Small IQR: Data is tightly packed around the median.

What to say:

"The IQR is narrow, suggesting that the middle 50% of values are closely clustered around the median, with low variability in that range."

4. Comparing Std and IQR
If std > IQR, it might suggest extreme values or outliers.

If IQR ≈ std, the spread is fairly even.

What to say:

"The standard deviation is much larger than the IQR, hinting at possible outliers or a long-tailed distribution."

Moderate Correlations (|r| ≈ 0.3–0.7)

Positive Correlation (r > 0)
Meaning: As one variable increases, the other also increases.

Implication: The variables move in the same direction.

Negative Correlation (r < 0)
Meaning: As one variable increases, the other decreases.

Implication: The variables move in opposite directions.

Zero Correlation (r ≈ 0)
Meaning: There is no consistent relationship between the two variables.

Implication: Changes in one variable do not predict changes in the other.
